```python
# 4.DELETE FROM TABLE
import mysql.connector as connector


# HERE WE ARE MAKING CLASS AND METHODS
class mydb:
    def __init__(self):
        self.con = connector.connect(host='localhost',
                                     user='root',
                                     port=3306,
                                     password='$ShivD37Code',
                                     database='main_sql')
      
    def fetch(self):
        q3 = "select * from demo;"
        fire = self.con.cursor()
        fire.execute(q3)
        for row in fire:
            print("USERID: ",row[0])
            print("USERNAME: ",row[1])
            print("USERADDRES: ",row[2])
            print()

    
    def delete(self,userid):
        q4 = "delete from demo where id= {}".format(userid)
        fire = self.con.cursor()
        fire.execute(q4)
        self.con.commit()
        print("DELETED SUCESSFULLY")
        
        
helpaer = mydb() 
print("BEFORE DELETION:   ")
helpaer.fetch()
helpaer.delete(4)
print("AFTER DELETION:   ")
helpaer.fetch()

      
```

```python
# 5.UPDATE IN TABLE
import mysql.connector as connector


# HERE WE ARE MAKING CLASS AND METHODS
class mydb:
    def __init__(self):
        self.con = connector.connect(host='localhost',
                                     user='root',
                                     port=3306,
                                     password='$ShivD37Code',
                                     database='main_sql')
      
    def fetch(self):
        q3 = "select * from demo;"
        fire = self.con.cursor()
        fire.execute(q3)
        for row in fire:
            print("USERID: ",row[0])
            print("USERNAME: ",row[1])
            print("USERADDRES: ",row[2])
            print()
            print()
        
    def update(self,col1_name,n_name,col2_name,n_add,userid):
        q5 = "update demo set {} = '{}',{} = '{}' where id={}; ".format(col1_name,n_name,col2_name,n_add,userid) 
        fire = self.con.cursor()
        fire.execute(q5)
        self.con.commit()
        print("UPDATED SUCCESFULLY")

    
helpaer = mydb() 
print("BEFORE UPADATING:--> \n")
helpaer.fetch()
helpaer.update('name','lakhaman','addres','Janakpuri',5)
print("AFTER UPADATING:--> \n")
helpaer.fetch()
```
